Двое рабочих, работая одновременно, выполнили всю работу за 5 дней. Если бы первый рабочий работал в 2
раза быстрее, а второй в 2 раза медленнее, то всю работу они выполнили бы за 4 дня. За сколько дней
выполнил бы всю работу первый рабочий, работая один?

```python
from sympy.abc import x,y
from sympy import Eq
from sympy import solve

sys_eq = [Eq(1 / x + 1 / y, 1/5), Eq(2 / x + 1 / (2 * y), 1 / 4)]
print(solve(sys_eq, [x,y]))
```

```python
from sympy.abc import x
from sympy import Eq
from sympy import solve

eq = Eq(2 / x + (1 / 5 - 1 / x) / 2, 1 / 4)
print(solve(eq, x))
```
